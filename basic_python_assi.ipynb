```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

A prime number is greater than 1 and divisible only by 1 and itself.


```python
import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print("Number 1:", num1)
print("Number 2:", num2)

user_answer = int(input("Enter the product: "))

if user_answer == num1 * num2:
    print("Correct answer")
else:
    print("Wrong answer. Correct product is:", num1 * num2)
```

Two random numbers are generated and the user's answer is checked.


```python
print("Squares of even numbers:")

for i in range(1, 11):
    if i % 2 == 0:
        print(i, ":", i * i)
```

Only even numbers are selected and their squares are printed.


```python
text = "This is a sample text. This text will be used to demonstrate the word counter."

words = text.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(word, ":", count)
```

The text is split into words and each word count is stored in a dictionary.


```python
def is_palindrome(text):
    text = text.lower()
    text = ''.join(char for char in text if char.isalnum())
    return text == text[::-1]

word = input("Enter a word: ")
print(is_palindrome(word))
```

The string is cleaned and compared with its reverse.
